```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("titanic.csv")
df.head()
```

```python
print(df.info())
print("\nStatistical Summary:\n")
print(df.describe())
print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
fig, ax = plt.subplots(2,2,figsize=(12,8))

sns.countplot(x='Survived', data=df, ax=ax[0,0])
ax[0,0].set_title("Survival Count")

sns.histplot(df['Age'].dropna(), bins=20, kde=True, ax=ax[0,1])
ax[0,1].set_title("Age Distribution")

sns.countplot(x='Sex', hue='Survived', data=df, ax=ax[1,0])
ax[1,0].set_title("Gender vs Survival")

sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', ax=ax[1,1])
ax[1,1].set_title("Correlation Heatmap")

plt.tight_layout()
plt.show()
```
